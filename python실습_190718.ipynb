```python
import matplotlib.pyplot as plt

df = pd.read_csv('Real_estate.csv')
df.baths.hist() # df.bahts.plot(kind='hist')
plt.show()
```

```python
df_baths = df[df.baths>=2]
df_baths.head(5)
```

```python
df_baths.price.hist()
plt.xlabel('price')
plt.ylabel('result')
plt.show()
```

```python
df_baths.boxplot(column='price')
plt.show()
```

```python
df_baths.boxplot(column='price',by='city',vert=False)
plt.yticks(fontsize=7)
plt.xlabel('price')
plt.ylabel('city')
plt.show()
```
